```python

test_cases = test_manager_data
if not test_cases:
    print("Không có test cases để xử lý.")

results = []

for test_case in test_cases:
    question = test_case["question"]
    expected_tools = test_case["expected_output"]
    
    result = await Runner.run(planner_agent, input=question)
    selected_tools = result.final_output.tools
    
    is_correct = set(selected_tools) == set(expected_tools)
    results.append({
        "question": question,
        "expected_tools": expected_tools,
        "selected_tools": selected_tools,
        "correct": is_correct
    })
    
    print(f"Question: {question}")
    print(f"Expected: {expected_tools}")
    print(f"Selected: {selected_tools}")
    print(f"Correct: {is_correct}\n")

correct_count = sum(1 for r in results if r["correct"])
accuracy = correct_count / len(results) * 100
print(f"Total test cases: {len(results)}")
print(f"Correct test cases: {correct_count}")
print(f"Accuracy: {accuracy:.2f}%")
```

```python
import json

with open("./synthetic_data/data/database_search_agent_test_cases_extended.json", "r", encoding="utf-8") as f:
    test_database_search_agent_data = json.load(f)
```

```python
test_database_search_agent_data[0]
```

```python
accurate_strict=[]
accurate_no_strict=[]
result_params = []
for entry in test_database_search_agent_data:
    result = await Runner.run(database_search_agent, input=entry["question"])
    tool = result.new_items[0].raw_item
    result_params.append({"id": entry["id"], "arguments": tool.arguments})
    if entry["expected_tool"] == "get_real_estate_posts" and tool.name == entry["expected_tool"]:
        accurate_no_strict.append(entry["id"])
    elif entry["expected_tool"] == "get_real_estate_posts_strict" and tool.name == entry["expected_tool"]:
        accurate_strict.append(entry["id"])
```

```python
def check_params_accuracy(expected_params, predicted_params):
    required_fields = [
        'estate_type', 'district', 'square', 'no_bedrooms',
        'no_bathrooms', 'price', 'price_per_square'
    ]
    for field in required_fields:
        if expected_params.get(field) != predicted_params.get(field):
            return False
    return True
def calculate_confusion_matrix(total_records=80, no_strict_records=40, strict_records=40):
    TP = len(accurate_no_strict)
    TN = len(accurate_strict)
    FN = no_strict_records - TP
    FP = strict_records - TN

    return {
        'TP': TP,
        'TN': TN,
        'FP': FP,
        'FN': FN
    }

def calculate_metrics(confusion_matrix):
    TP = confusion_matrix['TP']
    TN = confusion_matrix['TN']
    FP = confusion_matrix['FP']
    FN = confusion_matrix['FN']

    accuracy = (TP + TN) / (TP + TN + FP + FN) if (TP + TN + FP + FN) > 0 else 0

    precision_no_strict = TP / (TP + FP) if (TP + FP) > 0 else 0
    recall_no_strict = TP / (TP + FN) if (TP + FN) > 0 else 0
    f1_no_strict = (2 * precision_no_strict * recall_no_strict / (precision_no_strict + recall_no_strict)) if (precision_no_strict + recall_no_strict) > 0 else 0

    precision_strict = TN / (TN + FN) if (TN + FN) > 0 else 0
    recall_strict = TN / (TN + FP) if (TN + FP) > 0 else 0
    f1_strict = (2 * precision_strict * recall_strict / (precision_strict + recall_strict)) if (precision_strict + recall_strict) > 0 else 0

    return {
        'Accuracy': accuracy,
        'Precision (no_strict)': precision_no_strict,
        'Recall (no_strict)': recall_no_strict,
        'F1-Score (no_strict)': f1_no_strict,
        'Precision (strict)': precision_strict,
        'Recall (strict)': recall_strict,
        'F1-Score (strict)': f1_strict
    }

def calculate_params_accuracy(test_data, result_params):
    correct_params = 0
    result_params_dict = {item['id']: item['arguments'] for item in result_params}

    for entry in test_data:
        entry_id = entry['id']
        expected_params = entry['expected_params']
        predicted_params = result_params_dict.get(entry_id, {})
        
        predicted_params = json.loads(predicted_params)
        if check_params_accuracy(expected_params, predicted_params):
            correct_params += 1
        else:
            print(entry["question"])
            print(expected_params)
            print("=======================================================")
            print(predicted_params)
            print("\n\n\n\n")

    total_records = len(test_data)
    params_accuracy = correct_params / total_records if total_records > 0 else 0
    return params_accuracy
```

```python
confusion_matrix = calculate_confusion_matrix()
metrics = calculate_metrics(confusion_matrix)
params_accuracy = calculate_params_accuracy(test_database_search_agent_data, result_params)
```
